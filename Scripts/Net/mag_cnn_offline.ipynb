```python
import csv
import numpy as np
l=[]
with open('../Data/stm32.csv','rt') as f: 
    cr = csv.reader(f)
    for row in cr:
       row = list(map(float, row))
       l.append(row) 
nl = np.array(l)
import matplotlib.pyplot as plt
import scienceplots
import numpy as np

color_table = ['#192947','#4081A4','#88C0C1','#B2D5C9']

with plt.style.context(['science']):
    plt.rcParams['image.cmap'] = 'gray' #   style
    plt.rcParams['savefig.dpi'] = 400 #
    plt.rcParams['figure.dpi'] = 400 #
    plt.rcParams["legend.frameon"] = True
    plt.rcParams['text.usetex'] = False
    plt.rc('font',family='Times New Roman') 
    plt.plot(nl[:,3],linewidth=1,color=color_table[1],label='Training MAE')
    plt.grid()
    plt.xlabel('Samples')
    plt.ylim([0,4])
    plt.ylabel('Process Time(ms)')
    plt.show()
```

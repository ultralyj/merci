```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots

color_table = ['#192947','#4081A4','#88C0C1','#B2D5C9']

labels=['no weight','with weight']
xlabels = ['Train Loss','Train MAE','Valilate Loss','Valilate MAE']
x = np.linspace(1,12,4)  # the label locations
width=0.5
data = [[2.9577,0.4768],[1.2816,0.4768],[2.0965,4.7592],[1.0340,1.3416]]
data = np.array(data)

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x()+rect.get_width()/2.-0.3, 1.03*height, '%s' % height, size=10, family="Times new roman")
            
with plt.style.context(['science']):
    plt.rcParams['image.cmap'] = 'gray' # 设置 颜色 style
    plt.rcParams['savefig.dpi'] = 400 #图片像素
    plt.rcParams['figure.dpi'] = 400 #分辨率
    plt.rcParams["legend.frameon"] = True
    plt.rcParams['text.usetex'] = False
    plt.rc('font',family='Times New Roman') 

    b1 = plt.bar(x-0.5*width, data[:,0], width=width,color=color_table[0],zorder=2,label=labels[0],linewidth=0.5)
    autolabel(b1)

    b2 = plt.bar(x+0.5*width, data[:,1], width=width,color=color_table[1],zorder=2,label=labels[1],linewidth=0.5)
    autolabel(b2)
    
    # plt.bar(x, avrstd[:,1],width=width,color=color_table[1],zorder=2,label='y',linewidth=1)
    plt.xticks(x,labels=xlabels)
    plt.ylim(0,6)
    plt.grid()
    plt.legend()
plt.show()
    
```
